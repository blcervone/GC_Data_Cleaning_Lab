## Part 4: Mis-typed Columns/Broken Data


```python
# Question 1: find the columns where most of the data seems to be one type of data, and there seems to be a data error. In the markdown, after exploring in code, add a markdown cell discussing which columns have the data error. 
display(pima_dedupped_df)
pima_dedupped_df.info()
print(pima_dedupped_df['BloodPressure'].value_counts())
print(pima_dedupped_df['BMI'].value_counts())
display(pima_dedupped_df[pima_dedupped_df['BMI'] == 'Error'])
print(pima_dedupped_df['Outcome'].value_counts())
```

```python
# Question 2: Remove the rows with data errors - should remove 4 records from pima_dedupped_df
pima_remove_error_1_df = pima_dedupped_df[(pima_dedupped_df['BloodPressure'] != 'Error') & (pima_dedupped_df['Outcome'] != 'ERROR') & (pima_dedupped_df['BMI'] != 'Error')]
print(pima_remove_error_1_df['Insulin'].count())
```

```python
# Question 3: After the data is fixed in your columns change the columns to the correct type and save this as the pima_fixed_columns_df.
pima_fixed_columns_df = pima_remove_error_1_df.copy()
pima_fixed_columns_df['BloodPressure'] = pima_fixed_columns_df['BloodPressure'].astype('int')
pima_fixed_columns_df['BMI'] = pima_fixed_columns_df['BMI'].astype('float')

def convert_bool(x):
    print(x)
    if x == "FALSE":
        return False
    elif x == "TRUE":
        return True

pima_fixed_columns_df['Outcome'] = pima_fixed_columns_df['Outcome'].apply(convert_bool)

pima_fixed_columns_df['Outcome'] = pima_fixed_columns_df['Outcome'].astype('bool')
```

```python
# Question 4: Run pima_fixed_columns_df.info() to confirm the columns have changed.
pima_fixed_columns_df.info()
display(pima_fixed_columns_df)
```

## Part 5: Outlier Detection and Removal


```python
# Question 1: Print out the Outliers in each column in the pima_fixed_columns_df dataframe, use the IQR method of outlier detection.
def is_iqr_outlier(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)
    return ~column.between(lower_bound, upper_bound)

print("Pregnancies Outliers:")
display(pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["Pregnancies"])])
print("Glucose Outliers:")
display(pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["Glucose"])])
print("Blood Pressure Outliers:")
display(pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["BloodPressure"])])
print("Skin Thickness Outliers:")
display(pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["SkinThickness"])])
print("Insulin Outliers:")
display(pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["Insulin"])])
print("BMI Outliers:")
display(pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["BMI"])])
print("DiabetesPedigreeFunction Outliers:")
display(pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["DiabetesPedigreeFunction"])])
print("Age Outliers:")
display(pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["Age"])])


```

```python
# Question 2: Use loc to remove outliers in each of the columns that have outliers, save this as pima_outlier_removed_df.
pima_outlier_removed_df = pima_fixed_columns_df[is_iqr_outlier(pima_fixed_columns_df["Pregnancies"]) == False]
print(pima_outlier_removed_df['Insulin'].count())
pima_outlier_removed_df = pima_outlier_removed_df[is_iqr_outlier(pima_fixed_columns_df["BloodPressure"]) == False]
print(pima_outlier_removed_df['Insulin'].count())
pima_outlier_removed_df = pima_outlier_removed_df[is_iqr_outlier(pima_fixed_columns_df["Insulin"]) == False]
pima_outlier_removed_df['Insulin'].count()
#Should be 4 records less than pima_fixed_columns_df because 4 outliers were found
```
